```python
import pandas as pd
```

```python
#initialize the parameters
factors=["current ratio","quick ratio","asset-liability ratio"]
filename = 'debt_paying_ability_table.xls'
data = pd.read_excel(filename)
```

```python
x = data.iloc[:1000,2:5].as_matrix()
y = data.iloc[:1000,5].as_matrix()
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.linear_model import RandomizedLogisticRegression as RLR
rlr = RLR() #build RLR model, filtering the variables
rlr.fit(x,y) #train the model
print(rlr.scores_)#get the ratios
```

```python
lr = LR()
lr.fit(x,y)
print("The average accuracy is: %s" % lr.score(x,y))
```
